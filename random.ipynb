# Random Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
df = pd.DataFrame(np.random.rand(900,250) * 10000)

# Split dataframe into 3 classes with 300 samples each.
df['class'] = 0
df.loc[300:599, 'class'] = 1
df.loc[600:, 'class'] = 2

df.head()
```

## 5 Features Random Sample

```python
# Five Features
X = df.iloc[:, :5]
y = df['class']
```

### PCA

```python
X_transformed = PCA().fit_transform(X)
```

```python
# PCA Scatter Plot
plt.figure(figsize=[8,6])
p = sns.scatterplot(
    x=X_transformed[:,0],
    y=X_transformed[:,1],
    hue=y,
    palette=sns.color_palette("hls", 3),
    )

p.set(title="Random 5 Features PCA", xlabel="PC1", ylabel="PC2")
plt.show()
```

### LDA

```python
X_transformed = LinearDiscriminantAnalysis().fit_transform(X, y)
```

```python
# LDA Scatter Plot
plt.figure(figsize=[8,6])
p = sns.scatterplot(
    x=X_transformed[:,0],
    y=X_transformed[:,1],
    hue=y,
    palette=sns.color_palette("hls", 3),
    )

p.set(title="Random 5 Features LDA", xlabel="LD1", ylabel="LD2")
plt.show()
```
